```python
#spliting dataset
x=df.iloc[:,1:30].values
y=df.iloc[:,31].values
type(x)
y
```

```python
#spliting data into train & test 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
# models function
def models(x_train,y_train):
  #logistic_regession
  from sklearn.linear_model import LogisticRegression
  log =LogisticRegression(random_state=0)
  #Train the model using the training sets
  log.fit(x_train,y_train)
  #dicision tree
  from sklearn.tree import DecisionTreeClassifier
  tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
  #Train the model using the training sets
  tree.fit(x_train,y_train)
  #Random forest 
  from sklearn.ensemble import RandomForestClassifier
  forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
  #Train the model using the training sets
  forest.fit(x_train,y_train)
  #Import svm model
  from sklearn import svm
  #Create a svm Classifier
  clf = svm.SVC(kernel='linear') # Linear Kernel
  #Train the model using the training sets
  clf.fit(x_train, y_train)
  #print training accuracy

  print('[0]LogisticRegression trinaing accuracy ',log.score(x_train,y_train))
  print('[1]DecisionTree trinaing accuracy ',tree.score(x_train,y_train))
  print('[2]RandomForest accuracy ',forest.score(x_train,y_train))
  print('[3] SVM trinaing accuracy ',clf.score(x_train,y_train))
  return log,tree,forest,clf
```

```python
model=models(x_train,y_train)
```

```python
# test model accuracy on test data on confisution matrix 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
  print("modle",i)
  print(classification_report(y_test,model[i].predict(x_test)))
  print(accuracy_score(y_test,model[i].predict(x_test)))
  print()

```
